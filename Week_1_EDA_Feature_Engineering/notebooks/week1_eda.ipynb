# Week 1 — EDA & Feature Engineering
This notebook runs the full Week 1 pipeline: data loading, cleaning, EDA, feature engineering, chart generation, and export of the cleaned dataset.

```python
# Standard imports
import os
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
# Add workspace root to path so local modules import cleanly
repo_root = Path('..').resolve()
sys.path.append(str(repo_root))
print('Environment ready')
```

```python
# Import Week 1 processing modules
from Week_1_EDA_Feature_Engineering.src.data_cleaning import load_data, validate_df, missing_value_report, clean_orders, save_cleaned
from Week_1_EDA_Feature_Engineering.src.feature_engineering import create_features
from Week_1_EDA_Feature_Engineering.src.visualization import generate_all_charts
print('Modules imported')
```

```python
# Paths
DATA_PATH = '../data/ecommerce_orders.xlsx'
CHARTS_DIR = '../charts'
OUTPUT_CSV = '../outputs/cleaned_dataset.csv'
os.makedirs(os.path.dirname(OUTPUT_CSV), exist_ok=True)
os.makedirs(os.path.dirname(CHARTS_DIR), exist_ok=True)
```

```python
# Load dataset
df = load_data(DATA_PATH)
print('Loaded rows,cols:', df.shape)
display(df.head())
```

```python
# Validation summary
val = validate_df(df)
print('Shape:', val['shape'])
print('Columns count:', len(val['columns']))
print('Dtypes sample:', dict(list(val['dtypes'].items())[:8]))
```

```python
# Missing value report
mreport = missing_value_report(df)
display(mreport)
mreport.to_csv('../outputs/missing_value_report.csv')
print('Missing value report saved to ../outputs/missing_value_report.csv')
```

```python
# Clean data
cleaned = clean_orders(df)
print('After cleaning rows,cols:', cleaned.shape)
display(cleaned.head())
```

```python
# Feature engineering
fe = create_features(cleaned)
print('After features rows,cols:', fe.shape)
display(fe.head())
```

```python
# Generate charts and save to ../charts
generate_all_charts(fe, charts_dir=os.path.join('..', 'charts'))
print('Charts generated')
```

```python
# Save cleaned dataset
save_cleaned(fe, os.path.join('..', 'outputs', 'cleaned_dataset.csv'))
print('Cleaned dataset saved')
```
